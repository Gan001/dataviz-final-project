```python
from sqlalchemy import create_engine
from config3 import password
import pymysql
pymysql.install_as_MySQLdb()
import pandas as pd
import os
```

```python
rds_connection_string = f"root:{password}@localhost/JPdata"
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
JP_data_df = pd.read_csv(os.path.join("final_data06282019.csv")) 
```

```python
JP_data_df.head()
```

```python
JP_data_df=JP_data_df.drop(columns='Unnamed: 0')
JP_data_df.head()
```

```python
#Load df into table in mysql
JP_data_df.to_sql(name='jpdata_table', con=engine, if_exists='replace', index=False)
```

```python
#Preview table to verify load worked
pd.read_sql_query('select * from jpdata_table limit 10', con=engine)
```
